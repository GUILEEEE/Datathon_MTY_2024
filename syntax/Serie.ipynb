```python
# Herramientas matemáticas
import numpy as np

# Lectura de .csv a Dataframes
import pandas as pd

# Herramienta para hacer copias de Dataframes sin conflictos
import copy

# Graficos
import plotly.express as px
###############################################################
#####                       ¡WARNING!                       ###
############ Para asegurar compatibilidad entre          ######
############  Sistemas operativos comentar Linux      #########
############################################################### 
# LINUX:
dot = '.'
# WINDOWS:
#dot = ''
output_path = dot+'./output/'
```

```python
# importar Ventas_Vuelos
Ventas_2023 = pd.read_csv(output_path+'Ventas_Vuelos_2023.csv')
```

```python
# Extraemos las columnas de interés
Ventas_2023_v1 = copy.deepcopy(Ventas_2023[['STD','Bebidas Calientes', 'Botanas', 'Galletas', 'Lacteos','Licores', 'Perecederos', 'Refrescos', 'Sopas']])
Ventas_2023_v1['STD'] = pd.to_datetime(Ventas_2023_v1['STD'])
Ventas_2023_v1.set_index('STD',inplace=True)
Ventas_2023_v1.head()
```

```python
Ventas_2023_v1_day = Ventas_2023_v1.resample('D').sum()
```

```python
Ventas_2023_v1_day
```

```python
Ventas_2023_day =Ventas_2023_v1_day.reset_index()
```

```python
def plot_series(Serie):
    fig = px.line(Serie, x="STD", y=Serie.columns)

    fig.update_layout(
        title='Serie de tiempo de tipos de producto vendidos en 2023',
        xaxis_title="Fecha",
        yaxis_title="Número de productos",
        legend_title="Tipo",
        font=dict(size=15)
    )
    fig.show()
```

```python
plot_series(Ventas_2023_day)
```

```python
Ventas_2023_day.to_csv(output_path+'Ventas_2023_day.csv')
```

```python
Ventas_2023_v1_week = Ventas_2023_v1.resample('W-SUN').sum()
Ventas_2023_v1_week = Ventas_2023_v1_week.reset_index()
```

```python
plot_series(Ventas_2023_v1_week)
```
