# Solución con algoritmo greedy (sin restricción de pesos)

```python
def greedy_aleatorio_mejorado(iteraciones=10000, probabilidades=[]):
    mejor_evaluacion = float('inf')
    mejor_solucion = None
    for _ in range(iteraciones):
        solucion = np.random.randint(0, 21, size=98)
        evaluacion = evaluar_objetivo(solucion,probabilidades)
        if evaluacion < mejor_evaluacion:
            mejor_solucion = solucion
            mejor_evaluacion = evaluacion
        elif random.random() < 0.0001:
            mejor_solucion = solucion
            mejor_evaluacion = evaluacion
    return mejor_solucion, mejor_evaluacion

def evaluar_objetivo(num_productos,probabilidades):
    return np.dot(num_productos,probabilidades)


# Probabilidades simuladas
probabilidades = prob_datos.iloc[0]


solucion_optima, valor_objetivo = greedy_aleatorio_mejorado(10000,probabilidades)
print("Solución óptima:", solucion_optima)
print("Valor de la función objetivo:", valor_objetivo)
```

```python
import pulp

def mochila_con_articulos_repetidos(P, b, c):
    # Crear el problema de maximización
    prob = pulp.LpProblem("Mochila con Artículos Repetidos", pulp.LpMaximize)

    # Variables de decisión
    n = len(b)
    x = [pulp.LpVariable(f"x_{i}", cat=pulp.LpInteger, lowBound=0, upBound=1) for i in range(n)]

    # Función objetivo
    prob += pulp.lpSum(b[i] * x[i] for i in range(n))

    # Restricción de capacidad
    prob += pulp.lpSum(c[i] * x[i] for i in range(n)) <= P

    # Resolver el problema
    prob.solve()

    # Obtener la solución
    solucion = [(x[i].name, x[i].varValue) for i in range(n)]
    valor_total = pulp.value(prob.objective)

    return solucion, valor_total


```

# Estimación para horario de madrugada

```python
# Datos del problema
P = 343  # Capacidad máxima en kilos
b = prob_datos.iloc[0]  # Probabilidad de productos
c = pesos_estimados  # Pesos de productos

# Resolver el problema
solucion, valor_total = mochila_con_articulos_repetidos(P, b, c)

# Mostrar la solución
print("Solución:")
nombres_productos = prob_datos.columns
for i, (variable, valor) in enumerate(solucion):
    nombre_producto = nombres_productos[i]
    print(f"{nombre_producto} = {valor}")

print("Valor total:", valor_total)
```
