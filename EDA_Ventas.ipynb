## Productos más vendidos

```python
# Conteo de productos
ventas_df3 = Ventas_df2.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(10)

# Histograma
ventas_df3.plot.bar()
```

```python
# Conteo por tipo de productos
ventas_df3 = Ventas_df2.groupby('ProductType')['Quantity'].sum().sort_values(ascending=False)

# Histograma
ventas_df3.plot.bar()
```
